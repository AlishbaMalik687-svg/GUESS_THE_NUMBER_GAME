```python
import random

def choose_level():
    print("\n Choose Difficulty Level:")
    print("1. Easy (1 - 10)   | 5 Attempts")
    print("2. Medium (1 - 50) | 7 Attempts")
    print("3. Hard (1 - 100)  | 10 Attempts")

    while True:
        choice = input("Enter choice (1/2/3): ")
        if choice == "1":
            return 10, 5, "Easy"
        elif choice == "2":
            return 50, 7, "Medium"
        elif choice == "3":
            return 100, 10, "Hard"
        else:
            print("Invalid choice! Try again.")


def play_game(high_score):
    max_num, attempts, level_name = choose_level()
    number = random.randint(1, max_num)
    score = 0

    print(f"\nLevel: {level_name}")
    print(f"Guess the number between 1 and {max_num}")
    print(f"You have {attempts} attempts\n")

    for attempt in range(1, attempts + 1):
        try:
            guess = int(input(f"Attempt {attempt}: "))
        except:
            print(" Enter a valid number!")
            continue

        if guess == number:
            print(" Correct! You Win!")
            score = (attempts - attempt + 1) * 10
            print(f" Your Score: {score}")
            break
        elif guess < number:
            print("Too Low!")
        else:
            print("Too High!")

    else:
        print(f"\n❌ You Lost! The number was: {number}")

    # High score logic
    # if score > high_score:
    #     high_score = score
    #     print("New High Score!")

    # print(f" High Score: {high_score}")
    # return high_score


def main():
    print(" WELCOME TO GUESS THE NUMBER GAME ")
    high_score = 0

    while True:
        high_score = play_game(high_score)

        again = input("\nPlay again? (y/n): ").lower()
        if again != 'y':
            print("\n Thanks for playing!")
            break


if __name__ == "__main__":
    main()
```
